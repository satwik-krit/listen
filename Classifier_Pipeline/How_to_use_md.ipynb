### This file is guide on how to use the classifier

to better understand the working, lets see how the classifiers are used (workflow):-
1. Takes in raw numpy arrays
2. as the processor converts the .wav file into numpy arrays we give in the input in form of numpy arrays only for simplicity
3. we run the .py file inside the Classifier_Pipeline directory only - make sure all the 5 models are present
4. the models deactivate when not - saving up unecessary active RAM usage
5. The output is just simple text - the component and the type of component
6. the outputs are:-
  

*   pump00


*   pump02

*   pump04
*   pump06


*   valve00


*   valve02
*   valve04

*   fan00
*   fan02

*   fan04

*   fan06
*   slider00


*   slider02

*   slider06
